```python
#Import Required Libraries

import numpy as np  #to handle arrays and photo pixels
import matplotlib.pyplot as plt  #data visualization tool
import seaborn as sns  #statistical plotting 
import pandas as pd #data manipulation

%matplotlib inline
```

```python
#Load Dataset

#select columns

columns = ['SepalLengthCm','SepalWidthCm', 'PetalLengthCm','PetalWidthCm','Species']

#load the data
#data = sns.load_dataset("Iris") 
df = pd.read_csv('Iris.csv',names=columns,skiprows=[0])

df.head()
print(df.info())
```

```python
# draw lineplot 
#data = pd.read_csv('Iris.csv',names = columns)
sns.lineplot(x="SepalLengthCm", y="SepalWidthCm", data=df)

  
# setting the title using Matplotlib
plt.title('Sepal Length v/s Sepal Width')
```

```python
df.describe()  #description of dataset
```

```python
#Visualize the whole dataset
x_vars = ["SepalLengthCm", "SepalWidthCm"]
y_vars = ["PetalLengthCm", "PetalWidthCm"]
sns.set(style="ticks")
sns.pairplot(df, hue='Species', x_vars=x_vars, y_vars=y_vars, height=4, aspect=1.5)
plt.show()
```

```python
# # Drop the first row, which contains non-numeric labels
# df = df.iloc[1:, :]

# Convert the remaining data to NumPy arrays
data = df.values
X = data[:, 0:4].astype(float)  # Features (sepal length, sepal width, petal length, petal width)
Y = data[:, 4]    # Target variable (species labels)
print(X)
print(Y)



```

```python
# Calculate average of each features for all classes
Y_Data = np.array([np.average(X[:, i][Y==j].astype('float32')) 
for i in range (X.shape[1])
    for j in (np.unique(Y))])
Y_Data_reshaped = Y_Data.reshape(4, 3)
Y_Data_reshaped = np.swapaxes(Y_Data_reshaped, 0, 1)
X_axis = np.arange(len(columns)-1)
width = 0.25
```

```python
# Plot the average
plt.bar(X_axis, Y_Data_reshaped[0], width, label = 'Setosa')
plt.bar(X_axis+width, Y_Data_reshaped[1], width, label = 'Versicolour')
plt.bar(X_axis+width*2, Y_Data_reshaped[2], width, label = 'Virginica')
plt.xticks(X_axis, columns[:4])
plt.xlabel("Features")
plt.ylabel("Value in cm.")
plt.legend(bbox_to_anchor=(1.3,1))
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(X,Y, test_size=0.2)
print(x_train)
```

```python
print(y_train)
```

```python

print(x_test)
```

```python
print(y_test)
```

```python
#MODEL:  SVM (SUPPORT VECTOR MACHINE ALGORITHM)
```

```python
from sklearn.svm import SVC 
model_svc = SVC()
model_svc.fit(x_train,y_train) #model fitting
```

```python
prediction1 = model_svc.predict(x_test)
#To calculate accuracy use accuracy_score
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction1))
print(accuracy_score(y_test,prediction1)*100)
accuracy1 = accuracy_score(y_test,prediction1)
```

```python
for i in range(len(prediction1)):
      print(y_test[i],prediction1[i])
```

```python
X_new = np.array([[3, 2, 1, 0.2], [  4.9, 2.2, 3.8, 1.1 ], [  5.3, 2.5, 4.6, 1.9 ]])
#Prediction of the species from the input vector
prediction = model_svc.predict(X_new)
print("Prediction of Species: {}".format(prediction))
```

```python
#MODEL:  LR (LOGISTIC REGRESSION ALGORITHM)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression

model_LR=LogisticRegression(solver='lbfgs',max_iter=1000)
model_LR.fit(x_train,y_train)
```

```python
prediction2=model_LR.predict(x_test)

#To calculate accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction2))
print(accuracy_score(y_test,prediction2)*100)
accuracy2 = accuracy_score(y_test,prediction2)
```

```python
for i in range(len(prediction2)):
      print(y_test[i],prediction2[i])
```

```python
X_new = np.array([[3, 2, 1, 0.2], [  4.9, 2.2, 3.8, 1.1 ], [  5.3, 2.5, 4.6, 1.9 ]])
#Prediction of the species from the input vector
prediction2 = model_LR.predict(X_new)
print("Prediction of Species: {}".format(prediction2))
```

```python
#MODEL:  DTC(DECISION TREE CLASSIFIERS)
```

```python
#Decision Classifiers
from sklearn.tree import DecisionTreeClassifier

model_DTC=DecisionTreeClassifier()
model_DTC.fit(x_train,y_train)
```

```python
prediction3 = model_DTC.predict(x_test)

#To Calculate accuracy 
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction3))
print(accuracy_score(y_test,prediction3)*100)
accuracy3 = accuracy_score(y_test,prediction3)
```

```python
for i in range(len(prediction3)):
    print(y_test[i],prediction3[i])
```

```python
X_new = np.array([[3, 2, 1, 0.2], [  4.9, 2.2, 3.8, 1.1 ], [  5.3, 2.5, 4.6, 1.9 ]])
#Prediction of the species from the input vector
prediction3 = model_DTC.predict(X_new)
print("Prediction of Species: {}".format(prediction3))
```
